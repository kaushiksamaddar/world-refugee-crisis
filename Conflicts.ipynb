```python
# dependencies
import pandas as pd
import numpy as np
import os
from datetime import datetime
```

```python
#load conflicts csv
conflicts = os.path.join("1900-01-01-2018-02-20.csv")
```

```python
conflicts_df = pd.read_csv(conflicts)
```

```python
conflicts_df.head()
```

```python
#column headers for dataframe
list(conflicts_df)
```

```python
# find the oldest and newest entries of the data
conflicts_df['event_date'] = pd.to_datetime(conflicts_df['event_date'])
oldest_date = conflicts_df['event_date'].min()
newest_date = conflicts_df['event_date'].max()
print(oldest_date)
print(newest_date)
```

```python
# Identify the conflict types 
types = conflicts_df['event_type'].unique()
print(types)
```

```python
# What countries are included in the data
types = conflicts_df['country'].unique()
print(types)
```

```python
#limit columns to data useful for the project
conflicts = conflicts_df[['data_id','event_id_no_cnty','event_date','year','event_type','actor1','interaction','region','country','location','latitude','longitude','source','notes','fatalities','iso3']]
conflicts.head()
```

```python
# summary, avg, etc of all numerical data
conflicts.describe()
```

```python
#narrow down data to the most useful for the project
conflict_summary = conflicts[['event_date','event_type','region','country','location','latitude','longitude','fatalities','actor1']]
conflict_summary.head()
```

```python
# Rename columns
conflict_summary = conflict_summary.rename(columns={'event_date': 'Date', 'event_type': 'Conflict Type', 'region': 'Region', 'country': 'Country', 'location': 'Location', 'latitude': 'Latitude', 'longitude': 'Longitude', 'actor1': 'Actor'})
conflict_summary.head(10)
```

```python
#save to CSV
conflict_summary.to_csv('conflicts.csv')
```
